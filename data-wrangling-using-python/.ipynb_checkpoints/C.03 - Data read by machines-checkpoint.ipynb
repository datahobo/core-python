```python
# Now let's do some JSON.
import json
# use the standard call
json_data = open('data-text.json').read()
# now set data using the loads method - this is a STRING, not a FILE
data = json.loads(json_data)

# for item in data:
#    print item
```

```python
# XML data
from xml.etree import ElementTree as ET
## there are two other libraries - lxml and minidom

## tree for the XML
tree = ET.parse('data-text.xml')
## start at the root
root = tree.getroot()
## to get all the subelements of the root tree:
### list(root)
## There is an object named Data
data = root.find('Data')
## get all the subelements of the data tree
### list(data)
## All of them are called Observation
## get all the subelements of the Observations
### for observation in data:
###     for item in observation:
###         print item
###         print item.text -- works if there is text associated
###         print list(item) -- gets the children of the items
###         print item.attrib

# start with a blank dictionary
all_data = []

# loop through all the observations
for observation in data:
    # create an empty dictionary for the record
    record = {}
    for item in observation:
        
        lookup_key = item.attrib.keys()[0]
        # most of them are categorical, but one is numeric
        if lookup_key == 'Numeric':
            rec_key = 'NUMERIC'
            rec_value = item.attrib['Numeric']
        else:
            rec_key = item.attrib[lookup_key]
            rec_value = item.attrib['Code']
        
        record[rec_key] = rec_value
    
    # append that whole record to the dataset
    all_data.append(record)

#    print all_data
### for observation in data:
###     for item in observation:
```
