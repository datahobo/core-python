```python
# now we need to create a dictionary
# need to use indexing
## can index on lists using positive (2) and negative indexes (-2, second from the end)
## Python is zero-indexed
## slices take a subset of the list
## x[1:4] takes the first three items in a list
## x[5:] takes from the fifth item, to the end of the list
# updating the for loop to include a dictionary

data = {} # create an empty dictionary to store data

for i in xrange(14, sheet.nrows):
    row = sheet.row_values(i)
    country = row[1] # pulls the country name from the row
    data[country] = {} # adds the country name as the key to the dictionary
print data['Afghanistan'] #shows us the data for afghanistan, which is currently empty
```

```python
# now we need to create a dictionary and get the data

data = {} # create an empty dictionary to store data

for i in xrange(14, sheet.nrows):
    row = sheet.row_values(i)
    country = row[1] # pulls the country name from the row
    data[country] = {
        'child_labor': {
            'total': [],
            'male': [],
            'female': [],
        },
        'child_marriage': {
            'married_by_15': [],
            'married_by_18': [],
        }
    } 
print data['Afghanistan']
```

```python
# now we need to create a dictionary and get the data

data = {} # create an empty dictionary to store data

for i in xrange(14, sheet.nrows):
    row = sheet.row_values(i)
    country = row[1] # pulls the country name from the row
    data[country] = {
        'child_labor': {
            'total': [row[4], row[5]], # gets these from the zero-indexed cells
            'male': [row[6], row[7]],
            'female': [row[8], row[9]],
        },
        'child_marriage': {
            'married_by_15': [row[10], row[11]],
            'married_by_18': [row[12], row[13]],
        }
    }
    if country == 'Zimbabwe':
        break # last country we care about, based on visual inspection
print data['Afghanistan']
print data['Canada']
print data['Cambodia']
```
