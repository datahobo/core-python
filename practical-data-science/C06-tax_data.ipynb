```python
# use a generator to access the data one row at a time
# generators create functions that act as iterables.
## use them to perform filtering on demand and clean data as you read it
def dataset(path):
    with open(path, 'rU') as csvfile:
        # use with open to ensure that the CSV file is closed when the block is exited
        # replaces try, except, and finally statements
        reader = csv.DictReader(csvfile)
        for row in reader:
            yield row
```

```python
# which countries does this cover?
print set([row["Country"] for row in dataset(data_file)])
```

```python
# which years does this cover?
print min(set([int(row["Year"]) for row in dataset(data_file)]))
# using python list comprehension to generate a set
# concise statement that generates an iterable
## specify the output variable, along with the for keyword,
### and the iterable, to express the variable, along with an optional if condition
```

```python
# which years does this cover?
print max(set([int(row["Year"]) for row in dataset(data_file)]))
```

```python
# filter data for the US only. While it doesn't technically apply to this data set, the logic is good to write down
filter(lambda row: row["Country"] == "United States",
       dataset(data_file))
# filter creates a list from all of the values of a sequence or iterable that 
## make the function specified by the first parameter true
## we use an anonymous function to check it
```

```python
# time to start using matplotlib
import csv
import numpy as np
import matplotlib.pyplot as plt
def dataset(path, filter_field=None,
            filter_value=None):
    with open(path, 'rU') as csvfile:
        reader = csv.DictReader(csvfile)
        if filter_field:
            for row in filter(lambda row: row[filter_field]==filter_value, reader):
                yield row
#                else:
#                    for row in reader:
#                        yield row

                        
    def main(path):
        data = [(row["Year"], float(row["Average income per tax unit"]))
    for row in dataset(path, "Country", "United States")]

    width = 0.35
    ind   = np.arange(len(data))
    fig   = plt.figure()
    ax    = plt.subplot(111)

    ax.bar(ind, list(d[1] for d in data))
    ax.set_xticks(np.arange(0, len(data), 4))
    ax.set_xticklabels(list(d[0] for d in data) [0::4], rotation=45)
    ax.set_ylabel("Income in USD")
    plt.title("U.S. Average Income 1913 - 2012")

    plt.show()
    if __name__ == "__main__":
        main("us_income_trended.csv")
```
